```python
YAML_PATH = '/content/yolov10/safety_helmet_dataset/data.yaml'
EPOCHS = 20
IMG_SIZE = 640
BATCH_SIZE = 32
model.train(data=YAML_PATH ,
            epochs = EPOCHS ,
            batch = BATCH_SIZE ,
            imgsz = IMG_SIZE )
```

```python
# Đánh giá mô hình
TRAINED_MODEL_PATH = '/content/yolov10/runs/detect/train5/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```

```python
# Lưu mô hình sau khi huấn luyện
SAVED_MODEL_PATH = '/content/yolov10/trained_model.pt'
model.save(SAVED_MODEL_PATH)
```
